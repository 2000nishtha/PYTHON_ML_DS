# Basic Plotting with matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.arange(10)

y = x * 2

plt.plot(x, y)

plt.show()
```

```python
plt.plot([1, 3, 5, 7, 9], [1, 8, 3, 5, 1])

plt.show()
```

```python
# Labeling the plot

heights = np.array([160, 180, 146, 162, 184, 180])

weights = np.array([50, 78, 45, 51, 80, 60])

plt.plot(heights, weights, 'b+')

plt.xlabel('Heights (cm)')
plt.ylabel('Weights (kg)')
plt.title('Scatter Plot - Height & Weight')

plt.xlim([140, 200])
plt.ylim([40, 90])

plt.show()
```

```python
# Multiple plots in a single figure

x = np.arange(1, 21)
y_1 = np.log(x)
y_2 = np.exp(x)

# Initiating a figure
plt.figure(1)

# Creating a plot with 1 row and 2 cols

plt.subplot(121)                      # plt.subplot(1, 2, 1)
plt.title('y = log(x)')
plt.plot(x, y_1)

plt.subplot(122)
plt.title('y = exp(x)')
plt.plot(x, y_2)

plt.show()
```

1. Very useful to detect outliers.
2. Shows 25th, 50th and 75th percentiles.

Median-> 50th percentile<br />
Quantiles-> 25th, 50th, 75th, 100th percentile<br /><br />

75th percentile-> 75 percent of all the observations are less than 75th percentile.

Eg: CAT Rank

# Plotting with Seaborn

## 1 Univariate Analysis (for Numerical Variables)
i.e Analyzing one variable at a time.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Lets try to create random normal distribution with 100 data points.

x = np.random.normal(size=100)

print(x.shape)
```

```python
sns.distplot(x)

# by default-> sns.distplot(x, hist=True, kde=True)
```

### 1.1 Histogram

<b>REFERENCE: </b>https://statistics.laerd.com/statistical-guides/understanding-histograms.php

```python
sns.distplot(x, kde=False, rug=True)

# This is basically a frequency plot
# Try 'rug=False'
```

### 1.2 Kernel Density Estimation (1D Dimensionality plot)

<b>REFERENCE: </b>http://www.mvstat.net/tduong/research/seminars/seminar-2001-05/

```python
sns.distplot(x, hist=False, rug=True)
```

```python
sns.kdeplot(x, shade=True)
```

## 2 Bivariate Analysis (for Numerical Variables)

```python
mean, cor = [0, 1], [(1, 0.5), (0.5, 1)]

data = np.random.multivariate_normal(mean, cor, 1000)

df = pd.DataFrame(data, columns=['x', 'y'])

print(df.shape)
```

```python
df.describe()
```

### 2.1 Scatter Plot
<b>REFERENCE: </b>http://mste.illinois.edu/courses/ci330ms/youtsey/scatterinfo.html

```python
sns.jointplot(x='x', y='y', data=df, kind = 'scatter')
```

### 2.2 Hexbin Plots
<b>REFERENCE: </b>https://www.meccanismocomplesso.org/en/hexagonal-binning/

```python
sns.jointplot(x='x', y='y', data=df, kind='hex', color='k')
```

### 2.4 Pair Plot

```python
# available datasets in seaborn 
# https://github.com/mwaskom/seaborn-data/

iris = sns.load_dataset('iris')
sns.pairplot(iris)
```

```python
sns.pairplot(iris, hue='species')
```

### 3.1 Categorical Scatterplots - stripplot()

```python
sns.stripplot(x = 'day', y='total_bill', data=tips)
```

### 3.2 Swarm Plot

This function is similar to stripplot(), but the points are adjusted (only along the categorical axis) so that they don’t overlap. This gives a better representation of the distribution of values, but it does not scale well to large numbers of observations. This style of plot is sometimes called a “beeswarm”.

```python
sns.swarmplot(x='day', y='total_bill', data=tips)
```

### 3.3 Boxplot

```python
sns.boxplot(x='day', y='total_bill', data=tips)
```

```python
print(tips.loc[tips.time=='Lunch', :])
```

```python
tips.loc[tips.time=='Lunch', 'day'].value_counts()
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, hue='time', dodge=True)

# dodge -> When hue nesting is used, whether elements should be shifted along the categorical axis.
```

```python
tips['weekend'] = tips['day'].isin(['Sat', 'Sun'])

tips.head()
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, \
            hue='weekend', dodge=False)
```
